```python
import pandas as pd
import numpy as np
import re
from tabulate import tabulate
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from collections import Counter
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.metrics import average_precision_score
import sys
from sklearn.metrics import accuracy_score

orig_stdout = sys.stdout
f = open('out.txt', 'w')
sys.stdout = f

df = pd.read_csv('projectDataset_top_20_w_norm.csv')
#clf = RandomForestClassifier()

def biotran(x):
    if x == 'medicine':
        return 1
    if x == 'nan':
        return 2
    if x == 'science':
        return 3
    if x == 'nutritionalsciences':
        return 4
    if x == 'social sciences':
        return 5
    if x == 'life sciences':
        return 6
    if x == 'neurology':
        return 7
    if x == 'health sciences':
        return 8
    if x == 'neoplasms':
        return 9
    if x == 'physiology':
        return 10
    if x == 'metabolism':
        return 11
    if x == 'psychiatry':
        return 12
    if x == 'pediatrics':
        return 13
    if x == 'biology':
        return 14
    if x == 'sportsmedicine':
        return 15
    if x == 'publichealth':
        return 16
    if x == 'complementarytherapies':
        return 17
    if x == 'endocrinology':
        return 18
    if x == 'molecularbiology':
        return 19
    if x == 'physicalandrehabilitationmedicine':
        return 20

def tran_countnorm_3Class(x):
    if x <= .005:
        return int(1) #low
    if x > .005 and x <= .4:
        return int(2) #average
    if x > .4:
        return int(3) #high
    
def tran_countnorm_5Class(x):
    if x <= .000005:
        return int(1) #very low
    if x > .000005 and x <= .2:
        return int(2) #low
    if x > .2 and x <= .5:
        return int(3) #average
    if x > .5 and x <= .7:
        return int(4) #high
    if x > .7:
        return int(5) #very high

def classApplied_3(pred):
    list = []
    for item in pred:
        if item == 1:
            list.append('Low')
        if item == 2:
            list.append('Medium')
        if item == 3:
            list.append('High')
    return list
            
def classApplied_5(pred):
    list = []
    for item in pred:
        if item == 1:
            list.append('Very Low')
        if item == 2:
            list.append('Low')
        if item == 3:
            list.append('Avg')
        if item == 4:
            list.append('High')
        if item == 5:
            list.append('Very High')
    return list
    
def runIt(features, reactions, currSet):
    print(currSet)
    whichClass = 5;
    
    if whichClass == 3:    
        df['lovetextnorm'] = df['norm_love'].apply(tran_countnorm_3Class)
        df['hahatextnorm'] = df['norm_haha'].apply(tran_countnorm_3Class)
        df['wowtextnorm'] = df['norm_wow'].apply(tran_countnorm_3Class)
        df['sadtextnorm'] = df['norm_sad'].apply(tran_countnorm_3Class)
        df['angertextnorm'] = df['norm_anger'].apply(tran_countnorm_3Class)
    
    if whichClass == 5:    
        df['lovetextnorm'] = df['norm_love'].apply(tran_countnorm_5Class)
        df['hahatextnorm'] = df['norm_haha'].apply(tran_countnorm_5Class)
        df['wowtextnorm'] = df['norm_wow'].apply(tran_countnorm_5Class)
        df['sadtextnorm'] = df['norm_sad'].apply(tran_countnorm_5Class)
        df['angertextnorm'] = df['norm_anger'].apply(tran_countnorm_5Class)
     
 #   print(df.groupby('lovetextnorm')['lovetextnorm'].count())
 #   print(df.groupby('hahatextnorm')['hahatextnorm'].count())
 #   print(df.groupby('wowtextnorm')['wowtextnorm'].count())
 #   print(df.groupby('sadtextnorm')['sadtextnorm'].count())
 #   print(df.groupby('angertextnorm')['angertextnorm'].count())

    X = df.drop('journalname', axis=1)
   #X = X.drop('subjectofpaper', axis=1)
    X = X.drop('titleofpaper', axis=1)
    X = X.drop('typeofpaper', axis=1)
    X = X.drop('publish_date', axis=1)
    X = X.drop('latestPost', axis=1)
    X = X.drop('facebook_ID', axis=1)
    X = X.drop('DOI_ID', axis=1)
    X = X.drop('alt_id', axis=1)
    X = X.drop('poster_type_science_comm', axis=1)
    X = X.drop('poster_type_researcher', axis=1)
    X = X.drop('poster_type_public', axis=1)
    X = X.drop('poster_type_practitioner', axis=1)
    X = X.drop('altmetric_score_history_3month', axis=1)
    X = X.drop('altmetric_score_history_6month', axis=1)
    X = X.drop('altmetric_score_history_1year', axis=1)
    X = X.drop('citeulike_count', axis=1)
    X = X.drop('connotea_count', axis=1)
   #X = X.drop('twitter_count', axis=1)
    X = X.drop('wikipedia_count', axis=1)
    X = X.drop('blogs_count', axis=1)
    X = X.drop('news_count', axis=1)
    X = X.drop('googleplus_count', axis=1)

   #X = X.drop('mendeley_count', axis=1)
   #X = X.drop('facebook_count', axis=1)
    X = X.drop('qna_count', axis=1)
    X = X.drop('policy_count', axis=1)
    # X = X.drop('reddit_count', axis=1)
    X = X.drop('love', axis=1)
    X = X.drop('haha', axis=1)
    X = X.drop('wow', axis=1)
    X = X.drop('sad', axis=1)
    X = X.drop('anger', axis=1)
    X = X.drop('like', axis=1)

    X = X.drop('lovetextnorm', axis=1)
    X = X.drop('hahatextnorm', axis=1)
    X = X.drop('wowtextnorm', axis=1)
    X = X.drop('sadtextnorm', axis=1)
    X = X.drop('angertextnorm', axis=1)

   #X = X.drop('liketext', axis=1)
   #X = X.drop('', axis=1)
    X = X.drop('altmetric_score', axis=1)

    X = X.drop('norm_love', axis=1)
    X = X.drop('norm_haha', axis=1)
    X = X.drop('norm_wow', axis=1)
    X = X.drop('norm_sad', axis=1)
    X = X.drop('norm_anger', axis=1)
    X = X.drop('sum', axis=1)
    
    for dropIt in features:
        X = X.drop(dropIt, axis=1)

    if reactions == 1:
        print("\n-----------------------------LOVE-----------------------------")
        y = df['lovetextnorm'] #predicting 84% accuracy
    if reactions == 2:
        print("\n-----------------------------HAHA-----------------------------")
        y = df['hahatextnorm'] #predicting 97% accuracy
    if reactions == 3:
        print("\n-----------------------------WOW-----------------------------")
        y = df['wowtextnorm'] #predicting 90% accuracy
    if reactions == 4:
        print("\n-----------------------------SAD-----------------------------")
        y = df['sadtextnorm'] #predicting 97% accuracy
    if reactions == 5:
        print("\n-----------------------------ANGER-----------------------------")
        y = df['angertextnorm'] #predicting 97% accuracy

    X['subjectofpaper'] = X['subjectofpaper'].apply(biotran)

    X

    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=4)

    random_forest = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=4)

    random_forest.fit(X_train, y_train)

    pred = random_forest.predict(X_test)

    if whichClass == 3:    
        list = classApplied_3(pred)
    if whichClass == 5:    
        list = classApplied_5(pred)

    acc = 100 * accuracy_score(y_test, pred)

    print('Random Forest Real Accuracy: ', round(acc, 2), '%.')

    if whichClass == 3:    
        listactual = classApplied_3(y_test)
    
    if whichClass == 5:    
        listactual = classApplied_5(y_test)

    ticker = 0
    '''
    print('Num: \t\tActual: \t\tPrediction:')
    for p,a in zip(list,listactual):
        print(str(ticker) + '\t\t' + a + '\t\t\t' + p)
        ticker = ticker + 1
    '''
    featureSetz = []
    for h,i in zip(X, random_forest.feature_importances_):
        print(h, i)
        featureSetz.append(h)
        
    print(featureSetz)
        
    #Create chart objects represented by each feature and its importance
    objects = featureSetz
    y_pos = np.arange(len(objects))
    performance = random_forest.feature_importances_

    #Create the chart
    plt.bar(y_pos, performance, align='center', alpha=0.8)
    plt.xticks(y_pos, objects,rotation=90)

    #Chart labeling
    plt.ylabel('Feature Importance')
    plt.xlabel('Feature')
    plt.title('Features Importances')

    plt.show()   
    
    RFC_cnf_matrix = confusion_matrix(y_test, pred)
   # print("\nConfusion Matrix: Random Forest Classifier")
   # print(RFC_cnf_matrix)
    print("Classification Report: Random Forest")
    print(classification_report(y_test, pred))
      
    #Bernoulii
    Bernouli_classifierModel = BernoulliNB()
    Bernouli_classifierModel.fit(X_train, y_train)
    Bernouli_y_predict = Bernouli_classifierModel.predict(X_test)
    Bernouli_cnf_matrix = confusion_matrix(y_test, Bernouli_y_predict)
   # print(Bernouli_cnf_matrix)
    print("Classification Report: Bernouli")
    print(classification_report(y_test, Bernouli_y_predict))
    
    #Decision Tree
    DecisionTree_classifierModel = DecisionTreeClassifier()
    DecisionTree_classifierModel.fit(X_train, y_train)
    DecisionTree_y_predict = DecisionTree_classifierModel.predict(X_test)
    DecisionTree_cnf_matrix = confusion_matrix(y_test, DecisionTree_y_predict)
   # print("Confusion Matrix: Decision Tree Classifier")
   # print(DecisionTree_cnf_matrix)
    print("Classification Report: Decision Tree")
    print(classification_report(y_test, DecisionTree_y_predict))
    
    print("Random Forest Accuracy: ", accuracy_score(y_test, pred))
    print("Naive Bayes Accuracy: ",accuracy_score(y_test, Bernouli_y_predict))
    print("Decision Tree Accuracy: ",accuracy_score(y_test, DecisionTree_y_predict))
    
featuresList = []

#featuresList.append(['titleofpaper'])
#featuresList.append(['facebook_count'])
#featuresList.append(['facebook_count', 'mendeley_count'])
#featuresList.append(['facebook_count', 'mendeley_count','twitter_count'])
featuresList.append(['facebook_count','mendeley_count','twitter_count', 'reddit_count'])
featuresList.append(['mendeley_count','twitter_count', 'reddit_count'])
featuresList.append(['facebook_count'])
featuresList.append([])

allPotentialFeatures = ['subjectofpaper', 'facebook_count', 'mendeley_count','twitter_count', 'reddit_count']

counter = 0

for featureSet in featuresList:
    counter=counter+1
    main_list = np.setdiff1d(allPotentialFeatures,featureSet)
    print ("\n\n***Features: ", main_list)
    for i in range(1,6):
        runIt(featureSet, i, counter)
      
sys.stdout = orig_stdout
f.close()

print("Complete")
```
